## Limpieza de datos

### Renombrar columnas


```python
df.rename(columns={
    'Transaction ID': 'Transaction_ID',
    'Price Per Unit': 'Price_Per_Unit',
    'Total Spent': 'Total_Spent',
    'Payment Method': 'Payment_Method',
    'Transaction Date': 'Transaction_Date'
}, inplace=True)
```

```python
df
```

### Convertir a NaN celdas vacías, UNKNOWN y ERROR

```python
a_nan = ['UNKNOWN', 'ERROR', '']
df = df.replace(a_nan, np.nan)
df
```

### Convertir a fecha la columna 'Transaction_Date'

```python
df['Transaction_Date']=pd.to_datetime(df['Transaction_Date'])
```

### Extraer año, mes y día de la columna 'Transaction_Date'

```python
df['Year']=df['Transaction_Date'].dt.year
df['Month']=df['Transaction_Date'].dt.month
df['Day']=df['Transaction_Date'].dt.day
df
```

### Convertir a dato numérico las columnas 'Quantity', 'Price_Per_Unit' y 'Total_Spent'

```python
columnas_numericas = ['Quantity', 'Price_Per_Unit', 'Total_Spent']
df[columnas_numericas] = df[columnas_numericas].astype(float)
```

### Rellenar valores faltantes en columnas 'Quantity', 'Price_Per_Unit' y 'Total_Spent'

```python

# Rellenar Quantity usando Total_Spent / Price_Per_Unit
df['Quantity'] = df['Quantity'].fillna(df['Total_Spent'] / df['Price_Per_Unit'])

# Rellenar Price_Per_Unit usando Total_Spent / Quantity
df['Price_Per_Unit'] = df['Price_Per_Unit'].fillna(df['Total_Spent'] / df['Quantity'])

# Rellenar Total_Spent usando Quantity * Price_Per_Unit
df['Total_Spent'] = df['Total_Spent'].fillna(df['Quantity'] * df['Price_Per_Unit'])
```

```python
df
```

### Rellenar con UNKNOWN valores faltantes en 'Payment_Method' y 'Location'

```python
df['Payment_Method'] = df['Payment_Method'].fillna('UNKNOWN')
df['Location'] = df['Location'].fillna('UNKNOWN')
df
```

```python
eliminar = ['Item', 'Transaction_Date', 'Year', 'Month', 'Day', 'Quantity', 'Price_Per_Unit', 'Total_Spent']

df = df.dropna(subset=eliminar)
df
```
